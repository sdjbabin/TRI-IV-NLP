```python
import re

# Preprocessing the text: converting to lowercase and removing punctuation
def pp_text(text):
    text = text.lower()
    words = re.findall(r'\b\w+\b', text)
    return words

# Getting the list of words from the paragraph
words = pp_text(paragraph)
```

# c) Frequency of each word and most and least frequent words

```python
from collections import Counter
import pandas as pd

# Calculating the frequency of each word
word_freq = Counter(words)

sr_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

most_fr = sr_freq[0]

least_fr = [word for word, count in sr_freq if count == sr_freq[-1][1]]

print(f"Word frequencies:\n{pd.DataFrame(sr_freq, columns=['Word', 'Frequency'])}\n")

print(f"Most frequent word: {most_fr}\n")

print(f"Least frequent words: {least_fr}\n")
```

```python
longest = max(words, key=len)

print(f"Longest word: {longest}")
```

```python
import re
from collections import Counter
import pandas as pd

# First begin by defining the paragraph
p = """
How does a gunman, in plain view of some spectators, manage to open fire on a former president at a campaign rally protected by federal and local law enforcement? That is the frightening and unanswered question facing the country — and security officials — after Donald Trump said he suffered a graze wound on his right ear while he was speaking in a field in Butler, Pennsylvania, and left the stage with a bloodied head. 
Secret Service personnel shot and killed the gunman, identified as Thomas Matthew Crooks of Bethel Park, Pennsylvania, immediately after he opened fire, the FBI said. A spectator was killed and two additional spectators were injured, but it was not immediately clear who shot them.
"""

def preprocess_text(t):
    """Converting text to lowercase and removing punctuation"""
    t = t.lower()
    words = re.findall(r'\b\w+\b', t)
    return words

def total_words(w):
    """Counting the total number of words"""
    return len(w)

def unique_words(w):
    """Counting the total number of unique words"""
    return len(set(w))

def word_frequencies(w):
    """Calculating and printing the frequency of each word, the most frequent word, and the least frequent words"""
    freq = Counter(w)
    sorted_freq = sorted(freq.items(), key=lambda x: x[1], reverse=True)
    most_freq_word = sorted_freq[0]
    least_freq_words = [word for word, count in sorted_freq if count == sorted_freq[-1][1]]
    
    print(f"Word frequencies:\n{pd.DataFrame(sorted_freq, columns=['Word', 'Frequency'])}\n")
    print(f"Most frequent word: {most_freq_word}\n")
    print(f"Least frequent words: {least_freq_words}\n")

def longest_word(w):
    """Finding and printing the longest word"""
    long_word = max(w, key=len)
    print(f"Longest word: {long_word}")
```

```python
def main():
    w = preprocess_text(p)
    print("To begin with here's the paragraph\n", p)
    while True:
        print("\nChoose what do you want to find out:")
        print("1. Find total words")
        print("2. Find total unique words")
        print("3. Find word frequencies")
        print("4. Find longest word")
        print("5. Exit")
        
        choice = input("Enter your choice (1-5): ")
        
        if choice == '1':
            print(f"Total words: {total_words(w)}")
        elif choice == '2':
            print(f"Total unique words: {unique_words(w)}")
        elif choice == '3':
            word_frequencies(w)
        elif choice == '4':
            longest_word(w)
        elif choice == '5':
            print("Exiting... Bye :(")
            break
        else:
            print("Invalid choice. Please enter a number from 1 to 5.")

if __name__ == "__main__":
    main()
```

# 2. Regular Expressions

2.1 Write regular expressions for the following languages.

1. the set of all alphabetic strings;
2. the set of all lower case alphabetic strings ending in a b;
3. the set of all strings from the alphabet a,b such that each a is immediately preceded by and immediately followed by a b;

### Ans 1

^[a-zA-Z]+$


### Ans 2

^[a-z]*b$


### Ans 3

^(b*ab+)*b*$


```python
import re

regex1 = r"^[a-zA-Z]+$"
regex2 = r"^[a-z]*b$"
regex3 = r"^(b*ab+)*b*$"

def test_regex(word, pattern):
    if re.match(pattern, word):
        return True
    return False

sentence = input("Enter a sentence: ")

words = re.findall(r'\b\w+\b', sentence)

matches = {
    "regex1": [],
    "regex2": [],
    "regex3": []
}

for word in words:
    if test_regex(word, regex1):
        matches["regex1"].append(word)
    if test_regex(word, regex2):
        matches["regex2"].append(word)
    if test_regex(word, regex3):
        matches["regex3"].append(word)

print(f"Sentence: {sentence}")
print(f" All alphabetic strings: {matches['regex1']}\n")
print(f" Lowercase alphabetic strings ending in 'b': {matches['regex2']}\n")
print(f" Strings where each 'a' is immediately preceded and followed by 'b': {matches['regex3']}\n")
```

an alphabetic string separated from other words by whitespace, any relevant
punctuation, line breaks, and so forth.

```python
import re

regex1 = r"\b(\w+)\s+\1\b"
regex2 = r"^\d+.*\b[a-zA-Z]+\b$"
regex3 = r"\b(?:grotto\b.*\braven|raven\b.*\bgrotto)\b"
regex4 = r"^([A-Z][a-zA-Z]*)\b"

# Test strings
test_strings = [
    "Boom Boom Bumrah",
    "umm umm i dont know what to say",
    "the bug",
    "7 is my lucky number",
    "123 start middle end",
    "In the grotto, there is a raven.",
    "The raven visited the grotto.",
    "Grottos and ravens",
    "hello, world! This is a test."
]

def test_regex(test_strings, pattern):
    matches = []
    for string in test_strings:
        match = re.search(pattern, string)
        if match:
            matches.append(string)
    return matches

print("Regex1 (two consecutive repeated words):")
regex1_matches = test_regex(test_strings, regex1)
for match in regex1_matches:
    print(f" - {match}")

print("\nRegex2 (starts with an integer, ends with a word):")
regex2_matches = test_regex(test_strings, regex2)
for match in regex2_matches:
    print(f" - {match}")

print("\nRegex3 (contains both 'grotto' and 'raven'):")
regex3_matches = test_regex(test_strings, regex3)
for match in regex3_matches:
    print(f" - {match}")

print("\nRegex4 (captures the first word of a sentence):")
regex4_matches = test_regex(test_strings, regex4)
for match in regex4_matches:
    print(f" - {match}")
```
